```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean1 = 0
mean2 = 100
```

```python
numsamples = 100
numexps = 10000
```

```python
single_estimators = []
double_estimators = []
stars = []

for i in range(numexps):
    sa1 = np.random.normal(mean1,size=numsamples)
    sa2 = np.random.normal(mean2,size=numsamples)

    sb1 = np.random.normal(mean1,size=numsamples)
    sb2 = np.random.normal(mean2,size=numsamples)

    mua1 = np.mean(sa1)
    mua2 = np.mean(sa2)
    mub1 = np.mean(sb1)
    mub2 = np.mean(sb2)


    single_est = np.max([mua1,mua2])

    astar = np.argmax([mua1,mua2])
    bstar = np.argmax([mub1,mub2])

    if astar == bstar:
        stars.append(1)
    else:
        stars.append(0)

    mubs = [mub1,mub2]
    double_est = mubs[astar]

    single_estimators.append(single_est)
    double_estimators.append(double_est)
```

```python
print(stars)
print(np.sum(stars))
```

```python
plt.plot(single_estimators,label='single est')
plt.plot(double_estimators,label='double est')
plt.legend()
```

```python
print('single est mean',np.mean(single_estimators))
print('double est mean',np.mean(double_estimators))

print('single est std dev',np.std(single_estimators))
print('double est std dev',np.std(double_estimators))
```
